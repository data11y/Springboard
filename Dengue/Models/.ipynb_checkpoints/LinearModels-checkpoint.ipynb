```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.linear_model import BayesianRidge
```

```python
with open('..\data\df_iq.pkl','rb') as f:
    iq = pickle.load(f)

with open('..\data\df_sj.pkl','rb') as f:
    sj = pickle.load(f)
```

```python
def rmsle(ytrue, ypred):
    y = ytrue.to_numpy() if isinstance(ytrue, pd.DataFrame) else ytrue
    return np.mean(np.sqrt((np.power(np.log1p(ypred) - np.log1p(y), 2))/(len(y))))

def getscores(ytrue, ypred, log=False):
    """Print RMSE, MAE, R^2 scores"""
    scores = {}
    scores['MAE'] = mean_absolute_error(ytrue, ypred)
    if log is False:
        scores['RMSE'] = mean_squared_error(ytrue, ypred, squared= False) 
    else:
        scores['RMSLE'] = rmsle(ytrue, ypred)
    scores['R2'] = r2_score(ytrue, ypred)
        
    return scores
```

## Linear Regression for month only

```python
#by month
iq['month'] = iq.index.month
sj['month'] = sj.index.month
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

plt.figure(figsize= (8,6))
plt.subplot(211)
iq.groupby('month')['total_cases'].sum().plot(kind= 'bar')
plt.xticks(np.arange(0, 12),months)
plt.title('Total Cases by month, Iquitos, Peru')

plt.subplot(212)
sj.groupby('month')['total_cases'].sum().plot(kind= 'bar')
plt.xticks(np.arange(0, 12),months)
plt.title('Total Cases by month, San Juan, Puerto Rico')

plt.tight_layout();
```

```python
months_dict = dict(zip(np.arange(1, 13),months))

# dummy variables for month column, rename and drop last for overhead
dummy_months_sj = pd.get_dummies(sj.month)
dummy_months_sj.columns = dummy_months_sj.columns.map(months_dict)

dummy_months_iq = pd.get_dummies(iq.month)
dummy_months_iq.columns = dummy_months_iq.columns.map(months_dict)

dummy_months_iq.sample(5)
```

```python
split_sj = (sj.shape[0] //4) *3  #train size .75, test size .25

sj_monthXtrain = dummy_months_sj.iloc[:split_sj]
sj_monthXtest = dummy_months_sj.iloc[split_sj:]
sj_monthYtrain = sj['total_cases'].iloc[:split_sj]
sj_monthYtest = sj['total_cases'].iloc[split_sj:]

split_iq = (iq.shape[0] //4) *3
iq_monthXtrain = dummy_months_iq.iloc[:split_iq]
iq_monthXtest = dummy_months_iq.iloc[split_iq:]
iq_monthYtrain = iq['total_cases'].iloc[:split_iq]
iq_monthYtest = iq['total_cases'].iloc[split_iq:]
```

```python
#SJ
sj_month_lr = LinearRegression()
sj_month_lr.fit(sj_monthXtrain, sj_monthYtrain)
sj_month_pred = sj_month_lr.predict(sj_monthXtest)
print("Metrics for San Juan ~ months only:")
display(getscores(sj_monthYtest, sj_month_pred))

#IQ
iq_month_lr = LinearRegression()
iq_month_lr.fit(iq_monthXtrain, iq_monthYtrain)
iq_month_pred = iq_month_lr.predict(iq_monthXtest)
print("Metrics for Iquitos ~ months only:")
display(getscores(iq_monthYtest, iq_month_pred))
```

San Juan

```python
plt.figure(figsize= (16,6))
sns.lineplot(x= sj.iloc[:split_sj].index, y= sj['total_cases'].iloc[:split_sj], color= 'b', label= 'Train set')
sns.lineplot(x= sj_monthYtest.index, y= sj_monthYtest, color= 'g', label= 'Test Set Values')
sns.lineplot(x= sj_monthYtest.index, y= sj_month_pred, color= 'r', lw= 3, label= 'Predicted Values')
plt.title("Predicting Dengue Severity on month alone in San Juan", size= 17)
plt.xlabel("")
plt.xticks(size= 14)
plt.xlim(sj.index[0], sj.index[-1])
plt.ylabel('Number of cases', size= 13)
plt.legend(fontsize= 15);
```

Iquitos

```python
plt.figure(figsize= (16,6))
sns.lineplot(x= iq.iloc[:split_iq +1].index, y= iq['total_cases'].iloc[:split_iq +1], color= 'b', label= 'Train set')
sns.lineplot(x= iq_monthYtest.index, y= iq_monthYtest, color= 'g', label= 'True Values')
sns.lineplot(x= iq_monthYtest.index, y= iq_month_pred, color= 'r', lw= 3, label= 'Predicted Values')
plt.title("Predicting Dengue Severity on month alone in Iquitos", size= 17)
plt.xlabel("")
plt.xticks(size= 14)
plt.xlim(iq.index[0], iq.index[-1])
plt.ylabel('Number of cases', size= 13)
plt.legend(fontsize= 15);
```

We can see that just going off of the month of year, our simple model doesn't do a good job at predicting the number of cases in either city. There must be other factors in play that influence case numbers besides the time of year.

* Outbreaks that are caused by external factors: natural influences, societal events, etc. can play a factor in how well our model predicts case numbers. These events are not accounted for here or easily predicted, especially in cases like hurricanes.

## Preprocessing

```python
# def to_c:
#     input: data
#         - DataFrame with columns of degrees Kelvin values
#     process:
#         - convert columns from Kelvin to Celsius by applying the formula C = K - 273.15
#         - renames columns of passed DataFrame by replacing "K" in column name to "C"
#     output:
#         - DataFrame with renamed columns with converted values
def to_c(data):
    data = data.apply(lambda x: x - 273.15)
    data = data.rename(lambda x: x.replace("k","c"), axis= 'columns')
    return data
```

```python
#columns that need to be converted from Kelvin to Celsius
cols_to_c = ['reanalysis_air_temp_k', 'reanalysis_avg_temp_k',
             'reanalysis_dew_point_temp_k', 'reanalysis_max_air_temp_k',
             'reanalysis_min_air_temp_k']

#get df with new features
sj_c_df = to_c(sj[cols_to_c])
iq_c_df = to_c(iq[cols_to_c])

#drop old features, merge processed features
sj_copy = sj.copy()
sj_c= sj_copy.drop(cols_to_c, axis= 1).merge(sj_c_df, on= 'week_start_date')

iq_copy = iq.copy()
iq_c = iq_copy.drop(cols_to_c, axis= 1).merge(iq_c_df, on= 'week_start_date')
```

```python
# drop 'month' column and merge with dummies from before
iq_c_dumm = iq_c.drop('month',axis= 1).merge(dummy_months_iq, on= 'week_start_date')
sj_c_dumm = sj_c.drop('month',axis= 1).merge(dummy_months_sj, on= 'week_start_date')
```

```python
#split dataset

#sj
sj_X = sj_c_dumm.drop(['total_cases', 'city'], axis= 1)
sj_Xtrain = sj_X.iloc[:split_sj]
sj_Xtest = sj_X.iloc[split_sj:]

sj_Ytrain = sj['total_cases'].iloc[:split_sj]
sj_Ytest = sj['total_cases'].iloc[split_sj:]

#iq
iq_X = iq_c_dumm.drop(['total_cases', 'city'], axis= 1)
iq_Xtrain = iq_X.iloc[:split_iq]
iq_Xtest = iq_X.iloc[split_iq:]

iq_Ytrain = iq['total_cases'].iloc[:split_iq]
iq_Ytest = iq['total_cases'].iloc[split_iq:]
```

#### Imputation

```python
# Imputation

# I found BayesianRidge to work well given speed among the methods I tested
# note that we imputed train/test sets seperately to avoid train set from influencing test set
bayes_iqXtrain = pd.DataFrame(IterativeImputer(random_state= 1, estimator= BayesianRidge()).fit_transform(iq_Xtrain), index= iq_Xtrain.index, columns= iq_Xtrain.columns)
bayes_iqXtest = pd.DataFrame(IterativeImputer(random_state= 1, estimator= BayesianRidge()).fit_transform(iq_Xtest), index= iq_Xtest.index, columns= iq_Xtest.columns)

bayes_sjXtrain = pd.DataFrame(IterativeImputer(random_state= 1, estimator= BayesianRidge()).fit_transform(sj_Xtrain), index= sj_Xtrain.index, columns= sj_Xtrain.columns)
bayes_sjXtest = pd.DataFrame(IterativeImputer(random_state= 1, estimator= BayesianRidge()).fit_transform(sj_Xtest), index= sj_Xtest.index, columns= sj_Xtest.columns)
```

### Linear Regression on BayesianRidge imputed data

```python
# With imputation done, drop datetime variables
iq_NoMonths = bayes_iqXtrain.drop([*months, 'year', 'weekofyear'], axis= 1)
iq_NoMonths_test = bayes_iqXtest.drop([*months, 'year', 'weekofyear'], axis= 1)
sj_NoMonths = bayes_sjXtrain.drop([*months, 'year', 'weekofyear'], axis= 1)
sj_NoMonths_test = bayes_sjXtest.drop([*months, 'year', 'weekofyear'], axis= 1)
```

```python
# Linear Regression on BayesianRidge imputed data

#IQ 
iq_bayesLR = LinearRegression()
iq_bayesLR.fit(iq_NoMonths, iq_Ytrain)
iq_bayes_pred = iq_bayesLR.predict(iq_NoMonths_test)
iq_bayesMAE = mean_absolute_error(iq_Ytest, iq_bayes_pred)

#SJ
sj_bayesLR = LinearRegression()
sj_bayesLR.fit(sj_NoMonths, sj_Ytrain)
sj_bayes_pred = sj_bayesLR.predict(sj_NoMonths_test)

print('Metrics for San Juan:')
display(getscores(sj_Ytest, sj_bayes_pred))
print('Metrics for Iquitos:')
display(getscores(iq_Ytest, iq_bayes_pred))
```

```python
ax = plt.figure(figsize= (12,8)).add_subplot(211)

ax.scatter(x= sj_Ytest, y= sj_bayes_pred, s= 10)
ax.set_xlabel("Actual Case Numbers", size= 12)
ax.set_ylabel("Predicted Case Numbers", size= 12)
plt.title('San Juan, Puerto Rico Linear Regression', size= 15)

ax2 = plt.figure(figsize= (12,8)).add_subplot(212)
ax2.scatter(x= iq_Ytest, y= iq_bayes_pred, s= 10)
ax2.set_ylabel("Predicted Case Numbers", size= 12)
ax2.set_xlabel("Actual Case Numbers", size= 12)
plt.title('Iquitos, Peru Linear Regression', size= 15)

plt.tight_layout(pad= 3);
```
