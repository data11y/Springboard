If we assume the reports are independant and the weeks records are not influenced by the previous weeks records, we can apply use the Centeral Limit Theorem to determine whether these sample data is representative of the whole population and if these differences are representative of the whole population.

We'll sample with replacement and find the mean of the bootstraps to try and estimate the population means.

```python
"""
Function: draw_reps(data, function, size=1000)
    This function builds an array of size size, 
    calling the bootstrap function each time, 
    with each element being a bootstrap of data with a function 
    applied.
    
Inputs:
    data: column(s) to perform bootstrap on
    function: function to perform in bootstrap step(ie. np.mean())
    
Return:
    array of size size, containing the bootstraps of data with
    function function applied.
"""

def draw_reps(data,function,size=1000):
    reps_array = np.empty(size)
    
    for i in range(size):
        reps_array[i] = bootstrap(data,function)
        
    return reps_array


#sample with replacement
def bootstrap(data,func):
    bs_sample = np.random.choice(data,len(data))
    return func(bs_sample)
```

```python
np.random.seed(32)
N_reps=10000
bs_avgtemp_sj = draw_reps(sub_sj.reanalysis_avg_temp_k, np.mean,N_reps)
bs_avgtemp_iq = draw_reps(sub_iq.reanalysis_avg_temp_k, np.mean,N_reps)
```

```python
plt.figure(figsize=(10,5))
plt.hist(bs_avgtemp_sj, bins=30, histtype='step', label='Puerto Rico')
plt.hist(bs_avgtemp_iq, 30, histtype='step', label='Peru')
plt.xlabel('mean average air temp (Bootstrap)', size=15)
plt.axvline(bs_avgtemp_sj.mean(),color='k', ls='--', label='Mean SJ: %.2f'%bs_avgtemp_sj.mean())
plt.axvline(bs_avgtemp_iq.mean(),color='k', ls='--', label='Mean IQ: %.2f'%bs_avgtemp_iq.mean())
plt.legend();
```

It seems that the two cities indeed have distinct average air tempurature means. We'll apply a t-test to validate this and see how significant these differences are.

```python
dof = (2*N_reps)-2 #degrees of freedom
var_sj = np.var(bs_avgtemp_sj)
var_iq = np.var(bs_avgtemp_iq)
s = np.sqrt((var_sj + var_iq)/2)
t = (bs_avgtemp_iq.mean() - bs_avgtemp_sj.mean())/(s*np.sqrt(2/N_reps))
p = 1 - stats.t.cdf(t, df=dof)
print("t = ", t)
print("p = ", p)
```

We can see that the means are indeed different from our t-score.

For this test, we will look at the station precipitation amount in mm, seeing if the two cities are infact different.

* **Null Hypothesis**  The precipitation in San Juan is on average equal to that of the average precipitation in Iquitos.
* **Alternate Hypothesis** The averages of precipitation in the two cities is not equal.

Assuming the Null Hypothesis, **$H_0$** is true:

Though our distribution of data of precipitation readings is what appears exponential, utilizing the Central Limit Theorem, we can resample the data that we have to draw conclusions of what the mean value of the population would be if we had multiple samples to get a better representation.

```python
"""
shifted means array structure:
        array - mean(array) + combined_mean


shift_means(data,amount):
    shift the mean of data to amount
============
Inputs
============
        data: data to shift
        amount: value to shift mean of data to    
Return:
============
    shifted: shifted array
    
"""
def shift_means(data, amount):
    shifted = data - np.mean(data) + amount
    return shifted
```

```python
obs_mean = np.mean(pd.concat([sub_sj.station_precip_mm, sub_iq.station_precip_mm]))
obs_diff = np.mean(sub_sj.station_precip_mm) - np.mean(sub_iq.station_precip_mm)
print('observed mean of total dataset: %.3f'%obs_mean)
print('observed difference of means: %.3f'%obs_diff)
```

```python
np.random.seed(32)

shiftedSJ_reps=draw_reps(shift_means(sub_sj.station_precip_mm, obs_mean), np.mean,10000)
shiftedIQ_reps=draw_reps(shift_means(sub_iq.station_precip_mm, obs_mean), np.mean, 10000)

shifted_diffs = shiftedSJ_reps - shiftedIQ_reps
```

```python
p = np.sum(shifted_diffs >= obs_diff) / len(shifted_diffs)
p
```

With such a large p value, we would fail to reject the null hypothesis.
