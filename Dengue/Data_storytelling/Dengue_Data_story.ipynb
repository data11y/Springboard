### Correcting weekofyear

```python
# some weeks had weekofyear as 53 out of 52
# it seems that week 53 was really week 1 of that year

print("Iquitos:") 
display(df_iq.query("weekofyear == 53").total_cases)
print("San Juan:")
display(df_sj.query("weekofyear == 53").total_cases)
```

```python
print("Iquitos:") 
display(df_iq.query("year == 2005")['weekofyear'].head(3), df_iq.query("year == 2005")['weekofyear'].tail(3))
```

We can see that the 53rd week is actually the first week of the year mislabeled. This is the case of all week 53 rows. And the years that have a 53rd week enty all start on the second week 01-08-year and dont have a week 52.

What this means is:
* we can set the 53rd week to be the first week
* we can increment the rest of the weeks by 1 to correct the mismatched time stamps

```python
sj_years = [1993, 1999, 2005]
iq_years = [2005, 2010]

for year in sj_years:
    df_sj['weekofyear'] = np.where(df_sj.year == year, df_sj.weekofyear +1, df_sj.weekofyear)
    
df_sj['weekofyear'] = np.where(df_sj.weekofyear > 52, 1, df_sj.weekofyear)
df_sj.sort_index(inplace= True)

for year in iq_years:
    df_iq['weekofyear'] = np.where(df_iq.year == year, df_iq.weekofyear +1, df_iq.weekofyear)
    
df_iq['weekofyear'] = np.where(df_iq.weekofyear > 52, 1, df_iq.weekofyear)
df_iq.sort_index(inplace= True)
```

```python
def switch(city):
    switch={'sj':df_sj, 'iq':df_iq}
    return switch[city]
```

```python
#plot annual pattern of feature per week, given city, year,feature
def yearly_pattern(city,year, feature):
    quarters=np.append(1,13*np.arange(1,5))
    quarter_labels=['Jan','Apr','Jul','Oct','Dec']
    df_pattern = switch(city)
    df_pattern = df_pattern[df_pattern.year == year]
    plt.bar(df_pattern.weekofyear, df_pattern[feature], label=year)
    plt.xticks(quarters,quarter_labels)
    plt.xlim([0,54])
    plt.xlabel("Quarter")
    plt.ylabel(feature)
```

In order to compare and visualize tempurature and weather data, we need to standardize them.

#### Missing data

```python
scaler = StandardScaler()

sj_scaled = scaler.fit_transform(df_sj[df_sj.columns[3:-1]])
df_sj_scaled = pd.DataFrame(sj_scaled, columns= df_sj.columns[3:-1], index= df_sj.index)
df_sj_scaled['weekofyear'] = df_sj['weekofyear']

iq_scaled = scaler.fit_transform(df_iq[df_iq.columns[3:-1]])
df_iq_scaled = pd.DataFrame(iq_scaled, columns= df_iq.columns[3:-1], index= df_iq.index)
df_iq_scaled['weekofyear'] = df_iq['weekofyear']
```

#### Now we'll look at the trends of our different data collections scaled

Looking at different climate measures on a weekly average

```python
precip_cols = ['precipitation_amt_mm', 'reanalysis_precip_amt_kg_per_m2', 'reanalysis_sat_precip_amt_mm', 'station_precip_mm']

temp_cols = ['reanalysis_air_temp_k', 'reanalysis_avg_temp_k', 'reanalysis_dew_point_temp_k', 'reanalysis_max_air_temp_k',
            'reanalysis_min_air_temp_k', 'reanalysis_tdtr_k', 'station_avg_temp_c', 'station_diur_temp_rng_c', 
            'station_max_temp_c', 'station_min_temp_c']

humidity_cols = ['reanalysis_relative_humidity_percent', 'reanalysis_specific_humidity_g_per_kg']


# Plotting different climate data for Iquitos, normalized weekly averages
fig = plt.figure(figsize= (16,10))
ax_p = fig.add_subplot(311)

for p in precip_cols:
    df_iq_scaled.groupby('weekofyear')[p].mean().plot(alpha= 0.5, ax= ax_p).legend(loc= 'center left', bbox_to_anchor= (1, 0.5), frameon= False)
ax_p.set_xlim(1,52)
ax_p.set_xlabel('')
ax_p.set_title("Normalized Precipitation, Iquitos")

ax_t = fig.add_subplot(312, sharex = ax_p)
for t in temp_cols:
    df_iq_scaled.groupby('weekofyear')[t].mean().plot(alpha= 0.4, ax = ax_t).legend(loc= 'center left', bbox_to_anchor= (1, 0.5), frameon= False)
ax_t.set_xlabel('')
ax_t.set_title("Normalized Tempurature, Iquitos")

ax_h = fig.add_subplot(313, sharex = ax_p)
for h in humidity_cols:
    df_iq_scaled.groupby('weekofyear')[h].mean().plot(alpha= 0.5, ax= ax_h).legend(loc= 'center left', bbox_to_anchor= (1, 0.5), frameon= False)
ax_h.set_xlabel('Week of year', size= 15)
plt.title("Normalized Humidity, Iquitos")
plt.tight_layout(pad= 3)
```

```python
# Plot different climate data for Scan Juan, weekly averages normalized
fig = plt.figure(figsize= (16,10))
ax_p = fig.add_subplot(311)

for p in precip_cols:
    df_sj_scaled.groupby('weekofyear')[p].mean().plot(alpha= 0.5, ax = ax_p).legend(loc= 'center left', bbox_to_anchor= (1, 0.5), frameon= False)
ax_p.set_xlim(1,52)
ax_p.set_xlabel('')
ax_p.set_title("Normalized Precipitation, San Juan")

ax_t = fig.add_subplot(312, sharex = ax_p)
for t in temp_cols:
    df_sj_scaled.groupby('weekofyear')[t].mean().plot(alpha= 0.4, ax= ax_t).legend(loc= 'center left', bbox_to_anchor= (1, 0.5), frameon= False)
ax_t.set_xlabel('')
ax_t.set_title("Normalized Tempurature, San Juan")

ax_h = fig.add_subplot(313, sharex = ax_p)
for h in humidity_cols:
    df_sj_scaled.groupby('weekofyear')[h].mean().plot(alpha= 0.5, ax= ax_h).legend(loc= 'center left', bbox_to_anchor= (1, 0.5), frameon= False)
ax_h.set_xlabel('Week of year', size= 15)
ax_h.set_title("Normalized Humidity, San Juan")
plt.tight_layout(pad= 3)
```

One of the reasons we normalize is to compare these different features that measure the same thing using different units. Otherwise we wouldn't be able to compare precipitation amount in $mm$ with precipitation amount in $\frac{kg}{m^2}$. 

Here we see the climate conditions in San Juan all tend to increase in the later parts of the year, from week 25 to 40, similar to the trend in cases in San Juan we saw earlier. 

Iquitos is more variable. The climate pattersn don't move as concisely or as tightly as that of San Juan. This variablilty reminds us of the variability in the the dengue cases pattern for Iquitos.
